#### Duration
**Conclusion**
- The data is highly positive skewed
- The data has no missing value
- The data contains live consert, podcast etc. that are acting as outliers(we may need to remove those)

```python
duration_df = new_df.copy()
```

```python
duration_df = duration_df.assign(duration_ms = duration_df['duration_ms'] / (60 * 10**3))
duration_df.rename(columns = {'duration_ms':'duration_m'}, inplace = True)
```

```python
duration_df['duration_m'].describe()
```
